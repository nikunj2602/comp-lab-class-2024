```python
# Import necessary libraries
import mdtraj as md
import matplotlib.pyplot as plt

# Load the trajectory and topology files
# Replace 'output.trr' and 'adp.gro' with the full paths if needed
traj = md.load('//home/ndv3235/comp-lab-class-2024/Week5-ParallelTempering/Inputs/T300/adp_exchange3temps.trr', top='/home/ndv3235/comp-lab-class-2024/Week5-ParallelTempering/Inputs/T300/adp_exchange3temps.gro')

# Compute phi and psi dihedral angles
phi_indices, phi_angles = md.compute_phi(traj)
psi_indices, psi_angles = md.compute_psi(traj)

# Create a time array in nanoseconds
time = traj.time / 1000  # Convert from ps to ns

# Plot phi angles over time
plt.figure(figsize=(12, 5))
plt.plot(time, phi_angles[:, 0], label='Phi')
plt.xlabel('Time (ns)')
plt.ylabel('Phi (radians)')
plt.title('Phi Dihedral Angle over Time')
plt.legend()
plt.grid()
plt.show()

# Plot psi angles over time
plt.figure(figsize=(12, 5))
plt.plot(time, psi_angles[:, 0], label='Psi')
plt.xlabel('Time (ns)')
plt.ylabel('Psi (radians)')
plt.title('Psi Dihedral Angle over Time')
plt.legend()
plt.grid()
plt.show()
```
