```python
import numpy as np
import matplotlib.pyplot as plt

# Define file paths and labels for each dataset
file_paths = [
    "/home/ndv3235/comp-lab-class-2024/Week8-LAMMPS-crystallization/Project/1lyz_Potential.xvg",
    "/home/ndv3235/comp-lab-class-2024/Week8-LAMMPS-crystallization/Project/Cys64_80 to Serine_Potential.xvg",
    "/home/ndv3235/comp-lab-class-2024/Week8-LAMMPS-crystallization/Project/cys80toSer80_Potential.xvg",
    "/home/ndv3235/comp-lab-class-2024/Week8-LAMMPS-crystallization/Project/Firat5PPPPP_Potential.xvg",
    "/home/ndv3235/comp-lab-class-2024/Week8-LAMMPS-crystallization/Project/First5P and last 10P_Potential.xvg"
]

labels = [
    "Native Lysozyme (1lyz)",
    "Mutant (Cys64_80 to Serine)",
    "Mutant (Cys80 to Ser80)",
    "Mutant (First5PPPPP)",
    "Mutant (First5P and Last10P)"
]

# Define colors and styles for each plot line
colors = ['#FF5733', '#33FF57', '#3357FF', '#FF33A1', '#33FFF5']
line_styles = ['-', '--', '-.', ':', '-']

# Initialize the plot
plt.figure(figsize=(12, 8))

# Loop through each file and plot the data
for i, file_path in enumerate(file_paths):
    # Load the data, skipping header lines starting with "@" or "#"
    data = np.loadtxt(file_path, comments=('@', '#'))
    time = data[:, 0]  # First column: time
    potential_energy = data[:, 1]  # Second column: potential energy
    
    # Plot each dataset with unique color and line style
    plt.plot(time, potential_energy, label=labels[i], color=colors[i], linestyle=line_styles[i], linewidth=2)

# Add titles and labels
plt.title("Potential Energy vs Time for Native Lysozyme and Mutants", fontsize=16, fontweight='bold')
plt.xlabel("Time (ps)", fontsize=14)
plt.ylabel("Potential Energy (kJ/mol)", fontsize=14)

# Customize the legend, grid, and aesthetics
plt.legend(title="Protein Variants", fontsize=12, title_fontsize='13')
plt.grid(visible=True, linestyle='--', color='gray', alpha=0.7)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import MaxNLocator

# Define file paths and labels for each dataset
file_paths = [
    "/home/ndv3235/comp-lab-class-2024/Week8-LAMMPS-crystallization/Project/1lyz_Potential.xvg",
    "/home/ndv3235/comp-lab-class-2024/Week8-LAMMPS-crystallization/Project/Cys64_80 to Serine_Potential.xvg",
    "/home/ndv3235/comp-lab-class-2024/Week8-LAMMPS-crystallization/Project/cys80toSer80_Potential.xvg",
    "/home/ndv3235/comp-lab-class-2024/Week8-LAMMPS-crystallization/Project/Firat5PPPPP_Potential.xvg",
    "/home/ndv3235/comp-lab-class-2024/Week8-LAMMPS-crystallization/Project/First5P and last 10P_Potential.xvg"
]

labels = [
    "Native Lysozyme (1lyz)",
    "Mutant (Cys64_80 to Serine)",
    "Mutant (Cys80 to Ser80)",
    "Mutant (First5PPPPP)",
    "Mutant (First5P and Last10P)"
]

# Define colors and styles for each plot line
colors = ['#FF5733', '#33FF57', '#3357FF', '#FF33A1', '#33FFF5']
line_styles = ['-', '--', '-.', ':', '-']

# Initialize the plot
plt.figure(figsize=(12, 8))

# Loop through each file and plot the data
for i, file_path in enumerate(file_paths):
    # Load the data, skipping header lines starting with "@" or "#"
    data = np.loadtxt(file_path, comments=('@', '#'))
    time = data[:, 0]  # First column: time
    potential_energy = data[:, 1]  # Second column: potential energy
    
    # Plot each dataset with unique color and line style, adding transparency and smooth lines
    plt.plot(time, potential_energy, label=labels[i], color=colors[i], linestyle=line_styles[i], 
             linewidth=2.5, alpha=0.85)

# Enhance titles, labels, and layout with specific styling
plt.title("Potential Energy vs Time for Native Lysozyme and Mutants", fontsize=18, fontweight='bold', pad=15)
plt.xlabel("Time (ps)", fontsize=16, labelpad=10)
plt.ylabel("Potential Energy (kJ/mol)", fontsize=16, labelpad=10)

# Use a grid with subtle styling
plt.grid(visible=True, which='major', linestyle='--', linewidth=0.5, color='gray', alpha=0.5)

# Customize ticks and axis limits
plt.xticks(fontsize=13)
plt.yticks(fontsize=13)
plt.gca().xaxis.set_major_locator(MaxNLocator(integer=True))

# Legend styling to place outside the plot area
plt.legend(title="Protein Variants", fontsize=12, title_fontsize='13', loc='upper left', bbox_to_anchor=(1.05, 1), 
           borderaxespad=0., frameon=False)

# Remove top and right spines for a clean look
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

# Tight layout to adjust spacing, and optimize for clarity
plt.tight_layout(rect=[0, 0, 0.85, 1])  # Leaves space on right for legend

# Show the plot
plt.show()
```
