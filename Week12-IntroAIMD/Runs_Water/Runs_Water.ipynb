```python
import numpy as np
import matplotlib.pyplot as plt
import os

def smooth_data(data, window_size=5):
    """
    Smooth the input data using a simple moving average filter.
    
    Parameters:
    - data: The input data to be smoothed.
    - window_size: The size of the moving window for averaging.
    
    Returns:
    - The smoothed data.
    """
    return np.convolve(data, np.ones(window_size)/window_size, mode='valid')

def plot_energy(filename):
    # Initialize an empty list to store data
    data = []

    # Read the file
    with open(filename, "r") as f:
        for line in f:
            # Skip lines that are empty or start with a comment (#)
            if line.strip() == "" or line.strip().startswith("#"):
                continue
            try:
                # Convert the line to a list of floats and add to the data list
                data.append(list(map(float, line.split())))
            except ValueError:
                # Skip lines that cannot be converted to numbers
                continue

    # Convert the data list to a NumPy array
    data = np.array(data)

    # Check if the data has the correct format
    if data.shape[1] < 3:  # Must have at least 3 columns (step, time, energy)
        raise ValueError("The file format is incorrect. At least 3 columns are required.")

    # Extract time and energy data
    time = data[:, 1]  # 2nd column: Time (fs)
    energy = data[:, 4]  # 5th column: Energy (eV)

    # Smooth the energy data to reduce noise
    smoothed_energy = smooth_data(energy, window_size=10)  # Adjust window size for more/less smoothing

    # Ensure the directory exists before saving the plot
    output_dir = "../Figures/Water"
    os.makedirs(output_dir, exist_ok=True)  # Create the directory if it doesn't exist

    # Plot the data
    plt.figure()
    plt.plot(time[:len(smoothed_energy)], smoothed_energy, label="Smoothed Potential Energy")
    plt.xlabel("Time (fs)")
    plt.ylabel("Energy (eV)")
    plt.title("Energy vs. Time (Smoothed)")
    plt.legend()
    plt.grid()

    # Save the plot as a PNG file
    plt.savefig(f"{output_dir}/smoothed_energy_vs_time.png")
    plt.show()

# Call the function with the full file path
plot_energy("/home/ndv3235/comp-lab-class-2024/Week12-IntroAIMD/Runs_Water/water_aimd_1ns_every0.5fs-1.ener")
```

```python
import os

def xyz_to_pdb(xyz_file, pdb_file, box_size=None):
    """
    Convert an XYZ file to a PDB file.
    
    Parameters:
    - xyz_file: str - Path to the input XYZ file
    - pdb_file: str - Path to the output PDB file
    - box_size: tuple - Box size (a, b, c) in Angstroms. Default is None.
    """
    # Ensure the output directory exists
    output_dir = os.path.dirname(pdb_file)
    if not os.path.exists(output_dir) and output_dir != '':
        os.makedirs(output_dir)

    with open(xyz_file, 'r') as xyz, open(pdb_file, 'w') as pdb:
        lines = xyz.readlines()
        
        # First line: number of atoms
        num_atoms = int(lines[0].strip())
        
        # Second line: box size information (optional)
        if box_size:
            pdb.write(f"CRYST1{box_size[0]:9.3f}{box_size[1]:9.3f}{box_size[2]:9.3f}  90.00  90.00  90.00 P 1           1\n")
        
        # Convert atom data
        pdb.write(f"TITLE     Converted from {xyz_file}\n")
        pdb.write("MODEL        1\n")
        
        for i, line in enumerate(lines[2:2+num_atoms], start=1):
            parts = line.split()
            element, x, y, z = parts[0], float(parts[1]), float(parts[2]), float(parts[3])
            pdb.write(f"HETATM{i:5d}  {element:<2}  MOL     1    {x:8.3f}{y:8.3f}{z:8.3f}  1.00  0.00           {element:<2}\n")
        
        pdb.write("ENDMDL\n")
        pdb.write("END\n")
    print(f"Converted {xyz_file} to {pdb_file}")

# Example usage
xyz_to_pdb(
    xyz_file="/home/ndv3235/comp-lab-class-2024/Week12-IntroAIMD/Runs_Water/water_aimd_1ns_every0.5fs-pos-1.xyz",
    pdb_file="/home/ndv3235/comp-lab-class-2024/Week12-IntroAIMD/Runs_Water/water.pdb",  # Save inside the Runs_Water directory
    box_size=(12.42, 12.42, 12.42)  # Box size (Å)
)
```

```python
import mdtraj as md
import matplotlib.pyplot as plt
from itertools import combinations

def compute_pair_gr(traj_file, top_file, atom1, atom2, box_size, output_file):
    """
    Compute and visualize the radial distribution function (g(r)) for a specific atom pair.

    Parameters:
    - traj_file: str - Trajectory file (.pdb)
    - top_file: str - Structure file (.pdb)
    - atom1: str - Type of the first atom (e.g., "O", "H")
    - atom2: str - Type of the second atom (e.g., "H", "H")
    - box_size: float - Box size (in nm)
    - output_file: str - Name of the file to save the resulting graph (.png)
    """
    # Load the trajectory data
    traj = md.load(traj_file, top=top_file)

    # Select atoms
    indices1 = [atom.index for atom in traj.topology.atoms if atom.element.symbol == atom1]
    indices2 = [atom.index for atom in traj.topology.atoms if atom.element.symbol == atom2]

    # Create atom pairs
    pairs = [(i, j) for i in indices1 for j in indices2 if i != j]
    print(f"Number of {atom1}-{atom2} pairs: {len(pairs)}")

    # Compute radial distribution function (g(r))
    rdf, edges = md.compute_rdf(traj, pairs=pairs, r_range=(0, box_size / 2))
    r = edges[:len(rdf)]  # Calculate the r values (match rdf length)

    # Plot the graph
    plt.figure()
    plt.plot(rdf, r, label=f"g(r) for {atom1}-{atom2}")
    plt.xlabel("Distance r (nm)")
    plt.ylabel("g(r)")
    plt.title(f"Radial Distribution Function g(r) for {atom1}-{atom2}")
    plt.legend()
    plt.grid()
    plt.savefig(output_file)  # Save the plot as a PNG file
    plt.show()

# Example usage
box_size = 1.242  # Box size (in nm)
compute_pair_gr("/home/ndv3235/comp-lab-class-2024/Week12-IntroAIMD/Runs_Water/water.pdb", 
                "/home/ndv3235/comp-lab-class-2024/Week12-IntroAIMD/Runs_Water/water.pdb", 
                "O", "O", box_size, "/home/ndv3235/comp-lab-class-2024/Week12-IntroAIMD/Runs_Water/gr_oxygen_oxygen_water.png")

compute_pair_gr("/home/ndv3235/comp-lab-class-2024/Week12-IntroAIMD/Runs_Water/water.pdb", 
                "/home/ndv3235/comp-lab-class-2024/Week12-IntroAIMD/Runs_Water/water.pdb", 
                "O", "H", box_size, "/home/ndv3235/comp-lab-class-2024/Week12-IntroAIMD/Runs_Water/gr_oxygen_hydrogen_proton.png")
```
