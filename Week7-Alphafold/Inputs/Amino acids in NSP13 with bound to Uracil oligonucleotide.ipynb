```python
import mdtraj as md

traj = md.load('/home/ndv3235/comp-lab-class-2024/Week7-Alphafold/Inputs/6zsl (1).pdb')

# Select phosphate atoms (P) in RNA and oxygen atoms (O2, O4) in uracil
phosphate_atoms = traj.topology.select("element P and resname RNA")
uracil_oxygen_atoms = traj.topology.select("name O2 or name O4 and resname U")

# Select all protein atoms
protein_atoms = traj.topology.select("protein")

# Find all atoms within 0.5 nm of RNA phosphate atoms
close_phosphate_neighbors = md.compute_neighbors(traj, 0.5, phosphate_atoms)
# Filter for protein residues only
close_residues_phosphate = [protein_atoms[idx] for frame in close_phosphate_neighbors for idx in frame if idx in protein_atoms]

# Find all atoms within 0.5 nm of uracil oxygen atoms
close_uracil_neighbors = md.compute_neighbors(traj, 0.5, uracil_oxygen_atoms)
# Filter for protein residues only
close_residues_uracil = [protein_atoms[idx] for frame in close_uracil_neighbors for idx in frame if idx in protein_atoms]

# Print amino acids close to RNA phosphate atoms
print("Amino acids within 0.5 nm of RNA phosphate atoms:")
for idx in close_residues_phosphate:
    print(traj.topology.atom(idx).residue)

# Print amino acids close to uracil oxygen atoms
print("Amino acids within 0.5 nm of uracil oxygen atoms:")
for idx in close_residues_uracil:
    print(traj.topology.atom(idx).residue)
```

```python
Key Observations : 

Positively Charged Residues:

Lysine (LYS288) and Arginine (ARG443, ARG567) are positively charged, which facilitates strong electrostatic interactions with the negatively charged RNA phosphate groups. These residues likely play a crucial role in stabilizing the RNA-protein interaction.
Flexible and Polar Residues:

Glycine (GLY285, GLY287, GLY538) contributes flexibility, allowing the protein to adapt structurally to accommodate RNA.
Glutamine (GLN404, GLN537) and Threonine (THR286) can engage in hydrogen bonding with the oxygen atoms of uracil, further enhancing the stability of the RNA interaction.
Proline (PRO284):

Proline is typically linked to more rigid structures, providing stability to the binding region and potentially strengthening the RNA binding site.

Conclusion:
These amino acids are likely critical for RNA binding, as positively charged residues stabilize interactions with phosphates, while flexible and polar residues facilitate hydrogen bonding and structural adaptation. This pattern indicates the presence of a functional RNA-binding domain or region that is essential for the protein's function.
```
