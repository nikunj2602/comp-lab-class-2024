```python
import numpy as np
import matplotlib.pyplot as plt

# Specify the path to the file
file_path = '/home/ndv3235/comp-lab-class-2024/Week3-IntroToMD/Nikunjkumar_Week_3/potential.xvg'

# Load the data from potential.xvg
data = np.loadtxt(file_path, comments=['@', '#'])

# Extract time and potential energy values
time = data[:, 0]  # First column (time in ps)
potential_energy = data[:, 1]  # Second column (potential energy in kJ/mol)

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(time, potential_energy, color='blue', label='Potential Energy')
plt.title('Potential Energy vs. Energy Minimization Step')
plt.xlabel('Time (ps)')
plt.ylabel('Potential Energy (kJ/mol)')
plt.grid()
plt.legend()
plt.tight_layout()

# Show the plot
plt.show()
```
