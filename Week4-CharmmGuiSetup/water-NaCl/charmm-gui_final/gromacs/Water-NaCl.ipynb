```python
import MDAnalysis as mda
import numpy as np
import matplotlib.pyplot as plt
from MDAnalysis.analysis import align

# Load the trajectory and topology
u = mda.Universe('step5_5.tpr', 'md_center_waterNaCl.xtc')

# Select atoms for Water Oxygens, Sodium, and Oxygen
water_oxygens = u.select_atoms('name O and resname TIP3')  # Select water Oxygens
sodiums = u.select_atoms('name Na')  # Select Sodium ions
oxygens = u.select_atoms('name O and not resname TIP3')  # Select other Oxygens

# Function to calculate MSD
def calculate_msd(selection):
    msd = []
    num_frames = len(u.trajectory)
    initial_positions = selection.positions

    for ts in u.trajectory:
        current_positions = selection.positions
        squared_displacement = np.sum((current_positions - initial_positions) ** 2, axis=1)
        msd.append(np.mean(squared_displacement))

    return np.array(msd)

# Calculate MSD for each group
msd_water_oxygens = calculate_msd(water_oxygens)
msd_sodiums = calculate_msd(sodiums)
msd_oxygens = calculate_msd(oxygens)

# Time in nanoseconds (assuming your time step is 0.1 ns)
time = np.arange(len(msd_water_oxygens)) * 0.1  # Modify if your time step is different

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(time, msd_water_oxygens, label='Water Oxygens', color='blue')
plt.plot(time, msd_sodiums, label='Sodium', color='red')
plt.plot(time, msd_oxygens, label='Oxygen', color='green')
plt.xlabel('Time (ns)')
plt.ylabel('MSD (nm²)')
plt.title('Mean Squared Displacement vs Time')
plt.legend()
plt.grid()
plt.savefig('Figures/msd_plot.png')  # Save the plot
plt.show()

import numpy as np
import matplotlib.pyplot as plt
from MDAnalysis.analysis import align

# Load the trajectory and topology
u = mda.Universe('step5_5.tpr', 'md_center_waterNaCl.xtc')

# Select atoms for Water Oxygens, Sodium, and Oxygen
water_oxygens = u.select_atoms('name O and resname TIP3')  # Select water Oxygens
sodiums = u.select_atoms('name Na')  # Select Sodium ions
oxygens = u.select_atoms('name O and not resname TIP3')  # Select other Oxygens

# Function to calculate MSD
def calculate_msd(selection):
    msd = []
    num_frames = len(u.trajectory)
    initial_positions = selection.positions

    for ts in u.trajectory:
        current_positions = selection.positions
        squared_displacement = np.sum((current_positions - initial_positions) ** 2, axis=1)
        msd.append(np.mean(squared_displacement))

    return np.array(msd)

# Calculate MSD for each group
msd_water_oxygens = calculate_msd(water_oxygens)
msd_sodiums = calculate_msd(sodiums)
msd_oxygens = calculate_msd(oxygens)

# Time in nanoseconds (assuming your time step is 0.1 ns)
time = np.arange(len(msd_water_oxygens)) * 0.1  # Modify if your time step is different

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(time, msd_water_oxygens, label='Water Oxygens', color='blue')
plt.plot(time, msd_sodiums, label='Sodium', color='red')
plt.plot(time, msd_oxygens, label='Oxygen', color='green')
plt.xlabel('Time (ns)')
plt.ylabel('MSD (nm²)')
plt.title('Mean Squared Displacement vs Time')
plt.legend()
plt.grid()
plt.savefig('Figures/msd_plot.png')  # Save the plot
plt.show()
```
