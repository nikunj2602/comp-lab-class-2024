```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the file
file_path = 'Data/1hz3_T310.run.25000000.energy.xvg'

# Initialize lists to hold data
time = []
potential = []
kinetic_energy = []
total_energy = []
temperature = []
pressure = []
volume = []
```

```python
# Read and process the file
with open(file_path, 'r') as file:
    for line in file:
        if line.startswith(('#', '@')):
            continue  # Skip comments and metadata
        # Split the line into values
        values = line.split()
        time.append(float(values[0]) / 1000)  # Convert time from ps to ns
        potential.append(float(values[1]))
        kinetic_energy.append(float(values[2]))
        total_energy.append(float(values[3]))
        temperature.append(float(values[4]))
        pressure.append(float(values[5]))
        volume.append(float(values[6]))
        
```

```python
# Create a DataFrame for easier manipulation
data = pd.DataFrame({
    'Time (ns)': time,
    'Potential (kJ/mol)': potential,
    'Kinetic Energy (kJ/mol)': kinetic_energy,
    'Total Energy (kJ/mol)': total_energy,
    'Temperature (K)': temperature,
    'Pressure (bar)': pressure,
    'Volume (nm^3)': volume
```

```python
# Define a function to generate plots
def plot_quantity_vs_time(x, y, y_label):
    plt.figure(figsize=(8, 5))
    plt.plot(x, y, label=y_label)
    plt.xlabel('Time (ns)')
    plt.ylabel(y_label)
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
# Generate 6 separate plots
plot_quantity_vs_time(data['Time (ns)'], data['Potential (kJ/mol)'], 'Potential (kJ/mol)')
plot_quantity_vs_time(data['Time (ns)'], data['Kinetic Energy (kJ/mol)'], 'Kinetic Energy (kJ/mol)')
plot_quantity_vs_time(data['Time (ns)'], data['Total Energy (kJ/mol)'], 'Total Energy (kJ/mol)')
plot_quantity_vs_time(data['Time (ns)'], data['Temperature (K)'], 'Temperature (K)')
plot_quantity_vs_time(data['Time (ns)'], data['Pressure (bar)'], 'Pressure (bar)')
plot_quantity_vs_time(data['Time (ns)'], data['Volume (nm^3)'], 'Volume (nm^3)')
```

```python
import mdtraj as md
import numpy as np
import matplotlib.pyplot as plt
# Load the trajectory
trajectory = md.load('Data/1hz3_T310.stepid25000000.every100ps.nowater.xtc',
                     top='Data/1hz3_T310.start.nowater.gro')
# Get the indices of the first and last atom in the chain
atom_1 = 0  # Assuming first atom index
atom_last = trajectory.n_atoms - 1  # Assuming last atom index

# Compute the distance between the first and last atom at each time step
end_to_end_distance = md.compute_distances(trajectory, [[atom_1, atom_last]])
# Compute the radius of gyration at each time step
radius_of_gyration = md.compute_rg(trajectory)
# Get the time of each frame in nanoseconds (or convert as needed)
time = trajectory.time  # Time is usually in picoseconds, so divide by 1000 for nanoseconds

# Plot end-to-end distance and radius of gyration
plt.figure(figsize=(10, 6))
plt.plot(time, end_to_end_distance, label='End-to-End Distance')
plt.plot(time, radius_of_gyration, label='Radius of Gyration')
plt.xlabel('Time (ns)')
plt.ylabel('Distance (nm)')
plt.title('End-to-End Distance and Radius of Gyration vs Time')
plt.legend()
plt.show()
# Plot histogram for end-to-end distance
plt.figure(figsize=(10, 5))
plt.hist(end_to_end_distance, bins=50, density=True, alpha=0.7, label='End-to-End Distance')
plt.xlabel('End-to-End Distance (nm)')
plt.ylabel('Probability Density')
plt.title('Normalized Histogram of End-to-End Distance')
plt.legend()
plt.show()

# Plot histogram for radius of gyration
plt.figure(figsize=(10, 5))
plt.hist(radius_of_gyration, bins=50, density=True, alpha=0.7, label='Radius of Gyration')
plt.xlabel('Radius of Gyration (nm)')
plt.ylabel('Probability Density')
plt.title('Normalized Histogram of Radius of Gyration')
plt.legend()
plt.show()
```
