```python
import mdtraj as md
import numpy as np
import matplotlib.pyplot as plt
# Load the trajectory
trajectory = md.load('Data/1hz3_T310.stepid25000000.every100ps.nowater.xtc',
                     top='Data/1hz3_T310.start.nowater.gro')
# Get the indices of the first and last atom in the chain
atom_1 = 0  # Assuming first atom index
atom_last = trajectory.n_atoms - 1  # Assuming last atom index

# Compute the distance between the first and last atom at each time step
end_to_end_distance = md.compute_distances(trajectory, [[atom_1, atom_last]])
# Compute the radius of gyration at each time step
radius_of_gyration = md.compute_rg(trajectory)
# Get the time of each frame in nanoseconds (or convert as needed)
time = trajectory.time  # Time is usually in picoseconds, so divide by 1000 for nanoseconds

# Plot end-to-end distance and radius of gyration
plt.figure(figsize=(10, 6))
plt.plot(time, end_to_end_distance, label='End-to-End Distance')
plt.plot(time, radius_of_gyration, label='Radius of Gyration')
plt.xlabel('Time (ns)')
plt.ylabel('Distance (nm)')
plt.title('End-to-End Distance and Radius of Gyration vs Time')
plt.legend()
plt.show()
# Plot histogram for end-to-end distance
plt.figure(figsize=(10, 5))
plt.hist(end_to_end_distance, bins=50, density=True, alpha=0.7, label='End-to-End Distance')
plt.xlabel('End-to-End Distance (nm)')
plt.ylabel('Probability Density')
plt.title('Normalized Histogram of End-to-End Distance')
plt.legend()
plt.show()

# Plot histogram for radius of gyration
plt.figure(figsize=(10, 5))
plt.hist(radius_of_gyration, bins=50, density=True, alpha=0.7, label='Radius of Gyration')
plt.xlabel('Radius of Gyration (nm)')
plt.ylabel('Probability Density')
plt.title('Normalized Histogram of Radius of Gyration')
plt.legend()
plt.show()
```
