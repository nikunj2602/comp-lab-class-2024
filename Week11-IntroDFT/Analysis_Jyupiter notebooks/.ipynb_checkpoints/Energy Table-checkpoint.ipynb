```python
# smiles.csv 
file_path = 'smiles.csv'

# smiles.csv 
smiles_data = pd.read_csv(file_path)

PandasTools.AddMoleculeColumnToFrame(smiles_data, smilesCol='smiles')
smiles_data.head(1)
```

```python
radius=3
nBits=1024

ECFP6 = [AllChem.GetMorganFingerprintAsBitVect(x,radius=radius, nBits=nBits) for x in smiles_data['ROMol']]
```

```python
ECFP6[0]
```

```python
len(ECFP6[0])
```

```python
nBits = 1024
```

```python
from rdkit.Chem import AllChem

# Generate ECFP6 fingerprints (Morgan fingerprints with radius=3)
ECFP6 = [AllChem.GetMorganFingerprintAsBitVect(mol, radius=3, nBits=nBits) for mol in smiles_data.ROMol]
```

```python
ecfp6_bits = [list(fp) for fp in ECFP6]
```

```python
cfp6_name = [f'Bit_{i}' for i in range(nBits)]
df_morgan = pd.DataFrame(ecfp6_bits, index=smiles_data.smiles, columns=cfp6_name)

# Display the first row of the DataFrame
df_morgan.head(1)
```

```python
ref_smiles = 'COC(=O)c1c[nH]c2cc(OC(C)C)c(OC(C)C)cc2c1=O'
ref_mol = Chem.MolFromSmiles(ref_smiles)
ref_ECFP4_fps = AllChem.GetMorganFingerprintAsBitVect(ref_mol,2)
```

```python
ref_mol
```

```python
bulk_ECFP4_fps = [AllChem.GetMorganFingerprintAsBitVect(x,2) for x in smiles_data['ROMol']]
```

```python
from rdkit import DataStructs

similarity_efcp4 = [DataStructs.FingerprintSimilarity(ref_ECFP4_fps,x) for x in bulk_ECFP4_fps]
```

```python
smiles_data['Tanimoto_Similarity (ECFP4)'] = similarity_efcp4
PandasTools.FrameToGridImage(smiles_data.head(4), legendsCol="Tanimoto_Similarity (ECFP4)", molsPerRow=4)
```

```python
smiles_data = smiles_data.sort_values(['Tanimoto_Similarity (ECFP4)'], ascending=False)
PandasTools.FrameToGridImage(smiles_data.head(8), legendsCol="Tanimoto_Similarity (ECFP4)", molsPerRow=4)
```

```python
Analysis : 

Energy Comparison:

Aromatic molecules (Benzene, Furan) are more stable than non-aromatic ones (Cyclohexane).
Functionalized molecules like caffeine have significantly lower energy due to many stable covalent bonds.
```
